```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
data = pd.read_csv("/Users/disha/Desktop/Assignment01/data.csv", encoding='latin1')

x = data["no2"].dropna().values
```

```python
r = 102303705

a_r = 0.05 * (r % 7)
b_r = 0.3 * ((r % 5) + 1)
```

```python

z = x + a_r * np.sin(b_r * x)
```

```python
mu = np.mean(z)
```

```python
variance = np.var(z)
```

```python
lambda_param = 1 / (2 * variance)
```

```python
c = math.sqrt(lambda_param / math.pi)
```

```python
print("mu =", mu)
print("lambda =", lambda_param)
print("c =", c)
```

```python
z_min = np.min(z)
z_max = np.max(z)

z_curve = np.linspace(z_min, z_max, 500)
```

```python
pdf_curve = c * np.exp(-lambda_param * (z_curve - mu)**2)
```

```python
plt.hist(z, bins=30, density=True, alpha=0.6)
```

```python
plt.plot(z_curve, pdf_curve, linewidth=2)
```
